```
from pyspark.sql.functions import avg, col,lower,when,sum,year
from pyspark.sql.types import FloatType
```

```
department_df = spark.table("databricks_workspace.default.department")
employee_df = spark.table("databricks_workspace.default.employee")
employee_performance_df = spark.table("databricks_workspace.default.employee_performance")
```

### Remove Duplicate Rows

```
employee_performance_df = employee_performance_df.dropDuplicates(["Employee_ID", "Review_Date"])
display(employee_performance_df)
```

###  Handle Missing Values

```


# Fill missing Performance_Score with the average score
avg_performance_score = employee_performance_df.select(avg(col('Performance_Score'))).collect()[0][0]
employee_performance_df = employee_performance_df.na.fill({'Performance_Score': avg_performance_score})
```

### Standardize Formats

```
# Convert all Job Titles to lowercase
employee_df = employee_df.withColumn('Job_Title', lower(col('Job_Title')))
```

### Remove Unnecessary Columns

```
# Drop unnecessary columns
department_df = department_df.drop('Extra_Column1', 'Extra_Column2', 'Extra_Column3')
employee_df = employee_df.drop('Extra_Column1', 'Extra_Column2', 'Extra_Column3')
employee_performance_df = employee_performance_df.drop('Extra_Column1', 'Extra_Column2', 'Extra_Column3')
```

### Convert Data Types

```
# Ensure Performance_Score is a float
employee_performance_df = employee_performance_df.withColumn('Performance_Score', col('Performance_Score').cast(FloatType()))
```

### Handle Outliers

```
# Remove outliers in Performance_Score (example: scores outside 0-5 range)
employee_performance_df = employee_performance_df.filter((col('Performance_Score') >= 0) & (col('Performance_Score') <= 5))
```

### Validate Data Integrity

```
# Ensure all Employee_IDs in Fact_EmployeePerformance exist in Dim_Employee
employee_performance_df = employee_performance_df.join(employee_df, on='Employee_ID', how='inner')
```

```
display(employee_performance_df.select("Employee_ID", "Review_Date", "Performance_Score"))
```

### Consistent Categorical Data

```
# Standardize Gender column in Dim_Employee
employee_df = employee_df.withColumn(
    'Gender',
    when(col('Gender').isin('M', 'Male'), 'Male')
    .when(col('Gender').isin('F', 'Female'), 'Female')
    .otherwise('Other')
)
```

```
employee_performance_df.write.mode("overwrite").saveAsTable("default.employee_performance_v2")
```

```
average_performance_by_department = employee_performance_df \
    .groupBy('Department_ID') \
    .agg(avg('Performance_Score').alias('Average_Performance_Score'))
```

```
total_sales_by_department = employee_performance_df \
    .groupBy('Department_ID') \
    .agg(sum('Sales_Amount').alias('Total_Sales_Amount'))

high_performers_df = employee_performance_df.filter(col('Performance_Score') > 4.5)

# Sales Performance by Year
Performance_yr = employee_performance_df \
    .withColumn('Year', year(col('Review_Date'))) \
    .groupBy('Employee_ID') \
    .pivot('Year') \
    .agg(avg('Performance_Score').alias('Average_Performance_Score'))
```
